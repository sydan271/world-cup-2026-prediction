```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

load the results

```python
print("Loading 10 Million simulations...")
df = pd.read_csv('simulation_log.csv')

# Get EVERY team that won at least once, automatically sorted by most wins
all_teams = df['Champion'].value_counts().index.tolist()

# 1. Make the graph significantly wider and taller to fit the data and legend
plt.figure(figsize=(18, 10))

# 2. Generate a large color palette so every team gets a distinct color
colors = plt.cm.tab20(np.linspace(0, 1, len(all_teams)))

# 3. Plot every single team
for i, team in enumerate(all_teams):
    cumulative_wins = (df['Champion'] == team).cumsum()
    
    # Visual trick: Make the top 5 teams' lines thicker and fully opaque, 
    # while the underdog teams are slightly thinner and transparent so the graph isn't a solid blob.
    line_width = 2.5 if i < 5 else 1.2
    alpha_val = 1.0 if i < 5 else 0.6
    
    plt.plot(df['Iteration'], cumulative_wins, label=team, linewidth=line_width, alpha=alpha_val, color=colors[i])

# 4. Format the graph
plt.title("The Race to the Trophy: All Teams Over 1 Million Simulations", fontsize=20, fontweight='bold')
plt.xlabel("Simulation Iteration", fontsize=14)
plt.ylabel("Total Championships Won", fontsize=14)
plt.grid(True, linestyle='--', alpha=0.5)

# Prevent scientific notation on axes
plt.ticklabel_format(style='plain', axis='both')

# 5. Move the legend OUTSIDE the graph box so it doesn't block the data
plt.legend(title="Teams (Ranked by Wins)", 
           fontsize=10, 
           title_fontsize=12, 
           bbox_to_anchor=(1.02, 1), # Anchors legend outside the right edge
           loc='upper left', 
           ncol=2) # Splits the legend into 2 columns so it fits vertically

# This forces matplotlib to adjust the window size so the legend doesn't get cut off
plt.tight_layout()

plt.show()
```
